```python
x = pd.read_csv('final_x.csv').drop_duplicates().drop(columns = ['Unnamed: 0'])
pop = pd.read_csv('population_total_long.csv')

#left join x and pop
df_x = x.merge(pop, left_on = ["Country Name",'year'], right_on = ['Country Name', 'Year'], how = 'left').rename(columns = {'Count': 'population'}).drop(columns = ['Year'])
df_y = pd.read_csv('fianl_cleaned_y.csv').drop(columns = ['Unnamed: 0']).rename(columns = {'Country Code': 'country_code',
                                                                                          'Fiscal Year' : 'year',
                                                                                          'current_amount' : 'disbursement_amount'})
df_y = df_y[df_y['year'].isin([2015, 2016])]
```
