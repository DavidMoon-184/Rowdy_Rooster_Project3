#### cancer : Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease

```python
cancer = pd.read_csv('30-70cancerChdEtc.csv')
cancer.head()

#merge
cancer_merge = code.merge(cancer, left_on='Country Name', right_on='Location', how = 'left')
cancer_merge = cancer_merge[cancer_merge['Dim1'] == 'Both sexes'][['Country Code', 'Country Name', 'Period', 'First Tooltip']]
cancer_merge = cancer_merge.rename(columns = {'Period': 'year',
                                             'First Tooltip':'cancer'})
cancer_merge
```

```python
#cancer_merge
df_merge_1 = df.merge(cancer_merge, left_on= ['country_code', 'year'], right_on=['Country Code', 'year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_1

#air_death_merge
df_merge_2 = df_merge_1.merge(air_death_merge, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_2

#alchol_abuse
df_merge_3 = df_merge_2.merge(alchol_abuse, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_3

#basic_sanitation
df_merge_4 = df_merge_3.merge(basic_sanitation, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_4

#drinking_water
df_merge_5 = df_merge_4.merge(drinking_water, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_5

#hand_wash
df_merge_6 = df_merge_5.merge(hand_wash, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_6

#fuel_tech
df_merge_7 = df_merge_6.merge(fuel_tech, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_7

#crude_suicide
df_merge_8 = df_merge_7.merge(crude_suicide, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_8

#dentists
df_merge_9 = df_merge_8.merge(dentists, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_9

#tuberculosis
df_merge_10 = df_merge_9.merge(tuberculosis, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_10

#NTDs
df_merge_11 = df_merge_10.merge(NTDs, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_11

#doctors
df_merge_12 = df_merge_11.merge(doctors, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_12

#poisoning
df_merge_13 = df_merge_12.merge(poisoning, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_13

#unsafe_wash
df_merge_14 = df_merge_13.merge(unsafe_wash, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_14

#nursing
df_merge_15 = df_merge_14.merge(nursing, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_15

#pharmacists
df_merge_16 = df_merge_15.merge(pharmacists, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_16

#safely_sanitization
df_merge_17 = df_merge_16.merge(safely_sanitization, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_17

#safely_sanitization
df_merge_18 = df_merge_17.merge(tobacco, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_18

#UHC_coverage
df_merge_19 = df_merge_18.merge(UHC_coverage, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_19

#life_expectancy
df_merge_20 = df_merge_19.merge(life_expectancy, left_on= ['country_code','Country Name', 'year'], right_on=['Country Code','Country Name','year'], how = 'left').drop(['Country Code'], axis=1)
df_merge_20

#UHC_data_access
df_merge_21 = df_merge_20.merge(UHC_data_access, left_on= ['country_code','Country Name'], right_on=['Country Code','Country Name'], how = 'left').drop(['Country Code'], axis=1).drop(['year_y'], axis=1)
df_final = df_merge_21.rename(columns = {'year_x': 'year'})
df_final
```
