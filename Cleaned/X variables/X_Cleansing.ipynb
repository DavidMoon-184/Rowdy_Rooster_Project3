```python
import pandas as pd
```

```python
code = pd.read_csv('country_code.csv')
code
```

#### cancer : Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease

```python
cancer = pd.read_csv('30-70cancerChdEtc.csv')
cancer.head()
#merge
cancer_merge = code.merge(cancer, left_on='Country Name', right_on='Location', how = 'left')
cancer_merge = cancer_merge[cancer_merge['Dim1'] == 'Both sexes'][['Country Code', 'Country Name', 'Period', 'First Tooltip']]
cancer_merge = cancer_merge.rename(columns = {'Period': 'year',
                                             'First Tooltip':'cancer'})
cancer_merge
```

#### air_pollution_death : Ambient and household air pollution attributable death rate (per 100 000 population, age-standardized)

```python
air_death = pd.read_csv('airPollutionDeathRate.csv')
air_death.head()
#merge
air_death_merge = code.merge(air_death, left_on='Country Name', right_on='Location', how = 'left')
air_death_merge = air_death_merge[air_death_merge['Dim1'] == 'Both sexes'][['Country Code', 'Country Name', 'Period', 'First Tooltip']]
air_death_merge = air_death_merge.rename(columns = {'Period': 'year',
                                             'First Tooltip':'air_pollution_death'})
air_death_merge
```

```python
# for i,j in air_death_merge['air_pollution_death'].iterrows():
#     print(i,j)


cols = air_death_merge['air_pollution_death'].columns[air_death_merge['air_pollution_death'].dtypes.eq(object)]
air_death_merge[cols] = air_death_merge[cols].apply(pd.to_numeric, errors='coerce')
```
